<p>In this notebook, we will take a look at a dataset of The Office episodes, and try to understand how the popularity and quality of the series varied over time.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setting the plot(figure) size
plt.rcParams["figure.figsize"] = (12,6)
```

```python
# Read in the CSV as a DataFrame
office_df = pd.read_csv('office_episodes.csv')
```

```python
#check the first 10 rows of the dataset
office_df.head(10)
```

```python
#s setting the color schemes
colors = []
for ind,row in office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append("red")
    elif row['scaled_ratings'] >= 0.25 and row['scaled_ratings'] < 0.50 :
        colors.append("orange")
    elif row['scaled_ratings'] >= 0.50 and row['scaled_ratings'] < 0.75 :
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")   
```

```python
#sizes for the scatterplot
sizes = []
for i,row in office_df.iterrows():
    if row['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)
```

```python
fig1 = plt.figure()
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.scatter('episode_number','viewership_mil',data=office_df, c=colors, s=sizes)
plt.show()
```
