```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales = pd.read_csv('datasets/supermarket_sales.csv')
pd.set_option('display.max_rows', 100)
```

```python
# Figure size
plt.rcParams['figure.figsize'] = [8, 4]
```

Males vs females sales

```python
sales['Gender'].value_counts()
```

<font size='4'> Total sales by males vs females customers </font>

```python
# total sales, male vs females
sales_basedon_gender = sales.groupby('Gender').agg({'Total':'sum'})['Total']

# x values for the graph
gend = sales['Gender'].unique()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()

ax.bar(gend,sales_basedon_gender)

ax.set_yticks([i for i in range(0,200000,10000)])

plt.show()
```

```python
sales_by_product = sales.groupby('Product line').agg({'Total':'sum'})
```

```python
fig, ax = plt.subplots()

ax.bar(sales_by_product.index, sales_by_product['Total'])

ax.set_xticklabels(sales_by_product.index, rotation=90)
ax.set_title('Product vs Sale')
ax.set_xlabel('Product')
ax.set_ylabel('Sale')

plt.show()
```
