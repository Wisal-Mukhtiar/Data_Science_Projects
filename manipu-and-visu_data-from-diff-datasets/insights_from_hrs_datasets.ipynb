```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
pd.set_option('display.max_columns', 36)
```

```python
hrs_df = pd.read_csv('datasets/HRS.csv')
```

```python
males = hrs_df[hrs_df['Sex']=='M ']
females = hrs_df[hrs_df['Sex']=='F']
```

#### Engagement survey vs Salary (Gender wise)

```python
fig, ax = plt.subplots()

ax.scatter('EngagementSurvey','Salary', data=males, label='Males')
ax.scatter('EngagementSurvey','Salary', data=females, label='Females')
ax.set_xlabel('Engagement Survey')
ax.set_ylabel('Salary')
ax.set_title('Engagement Survey vs Salary')
ax.legend()
plt.show()
```

#### Engagement survey vs Salary (Department wise)

Specifying colors for different dataFrames

```python
dept1 = hrs_df[hrs_df['DeptID']==1]
dept2 = hrs_df[hrs_df['DeptID']==2]
dept3 = hrs_df[hrs_df['DeptID']==3]
dept4 = hrs_df[hrs_df['DeptID']==4]
dept5 = hrs_df[hrs_df['DeptID']==5]
```

```python
fig, ax = plt.subplots()

ax.scatter('EngagementSurvey','Salary', data=dept1, label='Department 1')
ax.scatter('EngagementSurvey','Salary', data=dept2, label='Department 2')
ax.scatter('EngagementSurvey','Salary', data=dept3, label='Department 3')
ax.scatter('EngagementSurvey','Salary', data=dept4,  label='Department 4')
ax.scatter('EngagementSurvey','Salary', data=dept5,  label='Department 5')
ax.set_xlabel('Engagement Survey')
ax.set_ylabel('Salary')
ax.set_title('Engagement Survey vs Salary')
ax.legend()
plt.show()
```

```python
values = hrs_df['RecruitmentSource'].value_counts().values
source = hrs_df['RecruitmentSource'].unique()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.bar(source, values)
ax.set_xticklabels(source, rotation=20)
ax.set_yticks(list(range(0,101,10))
ax.set_xlabel('Recruitment Source')
ax.set_ylabel('No of Recruitments')
ax.set_title('No of Recruitment vs Recruitments Sources')
fig.show()
```
